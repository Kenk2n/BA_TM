## 네트워크 이론 기초 

1. 네트워크를 어떻게 모델링 할 것인가
    + 네트워크를 표현하는 과정에서 가장 중요한 그래프, 행렬 
    + 관계의 유무만 표현하는 것은 이진 네트워크
    + 관계의 강도까지 표현하는 것은 가중 네트워크 (숫자와 굵기로 표현가능) 
2. 그래프의 표현 방법 
    + 인접 행렬 방식
        - 무방향 네트워크의 경우 대칭적인 행렬형태로 표현된다. 
    + 노드 리스트 방식
        - 인접한 링크가 있는 노드만 표기 
    + 엣지 리스트 방식 
        - 노드들을 연결한 링크들의 리스트 
3. 네트워크의 유형 
<img src="https://user-images.githubusercontent.com/52059483/93709407-e41d6380-fb78-11ea-9365-022e44c71ded.png" width="300" height="200">
방향성의 유무와 가중치 유무에 따라 크게 4가지 유형으로 나뉘어짐
4. 파당(clique)
네트워크가 서로 다 연결 되어있으면
5. 클러스터(cluster)
유사한 속성을 가지고 묶여 있는 노드들의 집합. 속성을 중심으로 노드가 묶여있으면 클러스터 
6. 에고 네트워크(ego network) 
분석하고자 하는 노드를 중심에 두고 그 노드와의 직접적 연결관계를 네트워크로 표현한 것 

## 소셜 네트워크 분석 지표

1. 네트워크의 크기 
    - 네트워크를 구성하는 노드의 수
2. 네트워크의 밀도 
    - 네트워크를 구성하는 노드들의 수
3. 포괄성
    - 전체 네트워크에서 연결되지 않은 것을 빼준다. 
    - 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현
4. 연결정도(Degree)
    - 해당 노드가 얼마나 영향력이 있고 활동 적인가 파악 가능
5. 연결강도(Strength)
    - 노드관의 관계의 강도 
6. 연결거리(Dstance)
    - 두 개 특정 노드 간에 가장 짧은 연결 경로 
    - 두 노드간 연결 거리가 짧을 수록 연결이 높아짐 
7. 상호성(Reciprocity)
    - 노드간 연결이 서로 쌍방향인 경우 
8. 군집화 계수(Clustering coefficeint)
    - 네트워크 상에서 특정 부분의 노드들이 밀접하게 연결되어 다른 부분보다 네트워크 밀도가 높아짐

```python
## 간단한 netowrkx 예제 
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 만들기
G = nx.DiGraph()

# 점(노드) 만들기
G.add_nodes_from([1, 2, 3, 4, 5])

# 선(엣지) 만들기
G.add_edges_from([(1, 2), (2, 1), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2,4), (4, 2),
                 (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3), (2, 3)])

# degree 만들기
degree = nx.degree(G)
print(degree)

# 그래프 그리기 
nx.draw(G,node_size=[500 + v[1]*500 for v in degree], with_labels=True)

```

```python

#mataplotlib은 networkx 에서 그림을 그릴 시 필요 

#symmetric network 그리기 
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
G_symmetric = nx.Graph()

G_symmetric.add_edge('Ahn',  'Sang')
G_symmetric.add_edge('Ahn',  'Marc')
G_symmetric.add_edge('Ahn',  'Ho')
G_symmetric.add_edge('Ahn',  'Chae')
G_symmetric.add_edge('Sang',   'Chae')
G_symmetric.add_edge('Chae','Marc')
G_symmetric.add_edge('George',  'John')
G_symmetric.add_edge('George',  'Ahn')

print(nx.info(G_symmetric))
plt.figure(figsize=(5,5))
nx.draw_networkx(G_symmetric);

```

```python
# coefficient 구하기
print(nx.clustering(G_symmetric,'Ahn'),"", nx.average_clustering(G_symmetric))
```

```python
# network distance 구하기 
nx.degree(G_symmetric, 'Chae')
```

```python
# Centrality 구하기 
print(nx.degree_centrality(G_symmetric))

# Eigenvector Centrality
print(nx.eigenvector_centrality(G_symmetric))

# Closeness Centrality
print(nx.closeness_centrality(G_symmetric))
```

```python
# Asymmetric Networks 그리기 
G_asymmetric = nx.DiGraph()
G_asymmetric.add_edge('A','B')
G_asymmetric.add_edge('A','D')
G_asymmetric.add_edge('C','A')
G_asymmetric.add_edge('D','E')
nx.spring_layout(G_asymmetric)
nx.draw_networkx(G_asymmetric)
```

```python
# Weighted Networks 그리기 
G_weighted = nx.Graph()

G_weighted.add_edge('Sangho',  'Laura',   weight=25)
G_weighted.add_edge('Sangho',  'Marc',    weight=8)
G_weighted.add_edge('Sangho',  'John',    weight=11)
G_weighted.add_edge('Sangho',  'SeongYun',weight=1)
G_weighted.add_edge('Laura',   'SeongYun',weight=1)
G_weighted.add_edge('SeongYun','Marc',    weight=1)
G_weighted.add_edge('George',  'John',    weight=8)
G_weighted.add_edge('George',  'Sangho',  weight=4)

elarge = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] > 8]
esmall = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] <= 8]

pos = nx.circular_layout(G_weighted)  # positions for all nodes

# 노드 그리고
nx.draw_networkx_nodes(G_weighted, pos, node_size=700)

# 엣지 그리고
nx.draw_networkx_edges(G_weighted, pos, edgelist=elarge,width=6)
nx.draw_networkx_edges(G_weighted, pos, edgelist=esmall,width=6, alpha=0.5, edge_color='b', style='dashed')

# 라벨 입력 
nx.draw_networkx_labels(G_weighted, pos, font_size=20, font_family='sans-serif')

plt.axis('off')
plt.show();
```

```python
G_fb = nx.read_edgelist("/Users/kenken/Downloads/facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
print(nx.info(G_fb))
```

```python
# plot 그리기 
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb);
```

```python
# 매개중심성 (betweeness centrality) 계산 가능 
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```
